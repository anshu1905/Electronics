```python
#importing all the libraries required for explanatory data analysis(EDA)
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#load the dataset(csv file) that is diabetes.csv into a variable data
data=pd.read_csv('diabetes.csv')
```

```python
#top 5 data values
data.head()
```

```python
#bottom 5 data values
data.tail()
```

```python
#represents the size of the data
#there are 768 rows and 9 columns
data.shape
```

```python
#provides statistical data of each column 
data.describe()
```

```python
#all the data labels
data.columns
```

```python
#information of data type ,null,count,etc.
data.info()
```

```python
#replace all nan value with mean of the columns
data.fillna(data.mean())
data.isnull().mean()
```

```python
#plot histogram(frequency distribution) of each label
data.hist(figsize=(15,15),color="#31CFB3",bins=5)
```

```python
#find the correlation between different labels
correlation=data.corr()
print(correlation)
#generate a heatmap for the same
sns.heatmap(correlation, annot=True,linewidths=.5,   xticklabels=correlation.columns, 
         yticklabels=correlation.columns, linecolor='white', cbar=True)
```

```python
#generate bar plot to get the relation between outcome and rest of the data labels

plt.figure(figsize=(15,15))

plt.subplot(3,3,1)
sns.barplot(data=data,y='Pregnancies',x='Outcome',hue='Outcome')

plt.subplot(3,3,2)
sns.barplot(data=data,y='Glucose',x='Outcome',hue='Outcome')

plt.subplot(3,3,3)
sns.barplot(data=data,y='BloodPressure',x='Outcome',hue='Outcome')

plt.subplot(3,3,4)
sns.barplot(data=data,y='SkinThickness',x='Outcome',hue='Outcome')

plt.subplot(3,3,5)
sns.barplot(data=data,y='Insulin',x='Outcome',hue='Outcome')

plt.subplot(3,3,6)
sns.barplot(data=data,y='BMI',x='Outcome',hue='Outcome')

plt.subplot(3,3,7)
sns.barplot(data=data,y='DiabetesPedigreeFunction',x='Outcome',hue='Outcome')

plt.subplot(3,3,8)
sns.barplot(data=data,y='Age',x='Outcome',hue='Outcome')
```
